```python
# import dependencies
import pandas as pd
import os
from datetime import datetime
```

```python

# list all the files to read from Aug 2022 to July 2023
files_2022_2023 = ["data/JC-202208-citibike-tripdata.csv",
                "data/JC-202209-citibike-tripdata.csv",
                "data/JC-202210-citibike-tripdata.csv",
                "data/JC-202211-citibike-tripdata.csv",
                "data/JC-202212-citibike-tripdata.csv",
                "data/JC-202301-citibike-tripdata.csv",
                "data/JC-202302-citibike-tripdata.csv",
                "data/JC-202303-citibike-tripdata.csv",
                "data/JC-202304-citibike-tripdata.csv",
                "data/JC-202305-citibike-tripdata.csv",
                "data/JC-202306-citibike-tripdata.csv",
                "data/JC-202307-citibike-tripdata.csv"]

# Create an empty list to store the dataframes
files_list = []

# loop through each of the files
for file in files_2022_2023:
    
    # read multiple files in a df
    df = pd.read_csv(file)
    
    # append the df to the list
    files_list.append(df)
    
df = pd.concat(files_list)
    
df.head()
```

```python
df.shape
```

```python
# trip duration
# difference = time1 - time2 

df['start_time'] = df['started_at'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S").strftime("%H:%M:%S"))
df['end_time'] = df['ended_at'].apply(lambda x: datetime.strptime(x, "%Y-%m-%d %H:%M:%S").strftime("%H:%M:%S"))

# Convert 'start_time' and 'end_time' columns to datetime objects
#df['start_time'] = pd.to_datetime(df['start_time'], format="%H:%M:%S")
#df['end_time'] = pd.to_datetime(df['end_time'], format="%H:%M:%S")
df.head()
```

```python
# Convert 'start_time' and 'end_time' columns to datetime objects
df['start_time'] = pd.to_datetime(df['start_time'], format="%H:%M:%S")
df['end_time'] = pd.to_datetime(df['end_time'], format="%H:%M:%S")

# Calculate the time duration by subtracting 'start_time' from 'end_time'
df['time_duration'] = df['end_time'] - df['start_time']
df.head()
```

```python
df.to_csv('data/citi_bike_2022_2023.csv', index=False)
```
